```python
with open('Example_Regex.txt','r', encoding ='utf16') as file:
    text = file.read()
```

```python
print(text) 
```

```python
import re 
re.findall('a{2,3}','aab daaaac aa')
```

```python
patterns = '[0-9]{4,5} ?[0-9]{3} ?[0-9]{3}'
print(re.findall(patterns, text))
```

```python
re.findall('(a|b|c)xz','bxz cxz a xz')
```

```python
re.findall('[abc]xz','bxz cxz a xz')
```

```python
re.findall('\w+@\w.+',text)
```

```python
for line in text.split('\n'):
    result = re.search('\w+@\w.+',line)
    if result:
        print(result.group())
```

```python
print(re.split("PHÒNG|KHOA|TRUNG TÂM|\n", text))
```

```python
import re

pattern = r'[0-9]{2}/[0-9]{2}/[0-9]{4}'

kq = re.findall(pattern, text)

print(kq)
```

```python
combine = r'(\d{2}/\d{2}/\d{4})\s+.+\s+(\w+@\w+.\w+.vn)'
kq1 = re.findall(combine, text)
print(kq1)
```
